```python
#Question number 1

#String Slicing in Python
'''String slicing is a method used to extract a portion (substring) of a string by specifying a range of indices.
Slicing allows you to take a "slice" of a string, from one index to another, with optional step values.

Syntax:
string[start:end:step]
'start': The index from where the slice begins (inclusive).
'end': The index where the slice ends (exclusive).
'step': The step value that determines the stride between characters in the slice.'''
#Examples of String Slicing:
string="hello, world"
#basic slicing
slice1=string[0:5]
print(slice1)
slice2=string[7:]
print(slice2)
#Omitting the Start or End:
slice3=string[:5]
print(slice3)
slice4=string[7:]
print(slice4)
#Negative Indexing:
slice5=string[-5:-1]
print(slice5)
#Using a Step Value:
slice6=string[::2]
print(slice6)
# Reversing a String Using Slicing:
slice7=string[::-1]
print(slice7)
#Extracting a Substring with Skipping:
slice8=string[0:5:2]
print(slice8)
```

```python
#QUESTION 2

'''Lists in Python are one of the most versatile and widely used data structures. They allow you to store and manipulate a collection of items in a flexible way. Here are the key features of lists in Python:

1. Ordered Collection
Definition: Lists maintain the order of elements as they were added.'''
my_list=[1,2,3,4,5]
print(my_list)

'''
2. Mutable
Definition: You can modify the elements of a list after it has been created, which means you can add, remove, or change items.'''
my_list.append(6)
print(my_list)
my_list.remove(3)
print(my_list)
my_list[0]=0
print(my_list)

'''
3. Dynamic Size
Definition: Lists can grow and shrink as needed. You can add or remove items without worrying about the size of the list.'''
fruits = ["apple", "banana"]
fruits.append("cherry")
print(fruits)  # Output: ['apple', 'banana', 'cherry']
fruits.remove("banana")
print(fruits)  # Output: ['apple', 'cherry']
'''
4. Heterogeneous Elements
Definition: Lists can contain elements of different data types. For example, you can mix integers, floats, strings, and even other lists.'''
mixed_list = [1, 2.5, "three", [4, 5]]
print(mixed_list)
'''
5. Nested Lists
Definition: Lists can contain other lists as elements, creating a nested structure.'''
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(nested_list)
'''
6. Indexing and Slicing
Definition: You can access individual elements in a list using indexing, and you can extract a sublist using slicing.'''
my_list = [1, 2, 3, 4, 5]
print(my_list[0])  # Output: 1
print(my_list[2:4])  # Output: [3, 4]
'''
7. List Comprehensions
Definition: List comprehensions provide a concise way to create lists by applying an expression to each item in an iterable.'''
squares = [x**2 for x in range(1, 6)]
print(squares)
'''
8. Built-in Methods
Definition: Lists come with a wide range of built-in methods for common operations, such as adding, removing, and sorting elements.
Common Methods:
append(): Adds an element to the end of the list.
insert(): Inserts an element at a specific position.
remove(): Removes the first occurrence of an element.
pop(): Removes and returns the last element (or an element at a specified position).
sort(): Sorts the list in ascending order.
reverse(): Reverses the order of elements in the list.'''

fruits = ["apple", "cherry", "banana"]
fruits.sort()
print(fruits)  # Output: ['apple', 'banana', 'cherry']
'''
9.Iteration
Definition: You can easily loop through elements of a list using loops like for or while.'''
for fruit in fruits:
    print(fruit)
```

```python
# QUESTION 3
'''In Python, lists are mutable, meaning you can access, modify, and delete their elements. Hereâ€™s how to do it, with examples:

1. Accessing Elements in a List
You can access elements in a list by using indexing and slicing.

Indexing: Use the index to access a specific element. Indexing starts from 0 (zero-based index).
Slicing: Access a range of elements by providing start and end indices.'''
fruits = ["apple", "banana", "cherry", "date"]

# Accessing a single element
print(fruits[1])    # Output: 'banana'

# Accessing the last element using negative index
print(fruits[-1])   # Output: 'date'

# Accessing a range of elements (slicing)
print(fruits[1:3])  # Output: ['banana', 'cherry']

# Accessing all elements after index 1
print(fruits[1:])   # Output: ['banana', 'cherry', 'date']
'''
2. Modifying Elements in a List
You can change the value of an element by assigning a new value to a specific index or a range of indices.'''

fruits = ["apple", "banana", "cherry", "date"]

# Modifying a single element
fruits[1] = "blueberry"
print(fruits)       # Output: ['apple', 'blueberry', 'cherry', 'date']

# Modifying a range of elements
fruits[0:2] = ["grape", "orange"]
print(fruits)       # Output: ['grape', 'orange', 'cherry', 'date']

'''
3. Deleting Elements from a List
You can delete elements from a list using several methods:

del statement: Removes an element or a slice of elements from the list.
remove(): Removes the first occurrence of a specific value.
pop(): Removes and returns an element at a specific index (or the last element if no index is specified).
clear(): Removes all elements from the list.'''
# Using del stement
fruits = ["apple", "banana", "cherry", "date"]

# Deleting an element by index
del fruits[2]
print(fruits)       # Output: ['apple', 'banana', 'date']

# Deleting a range of elements
del fruits[1:3]
print(fruits)       # Output: ['apple']

#remove statement
fruits = ["apple", "banana", "cherry", "banana"]

# Removing the first occurrence of 'banana'
fruits.remove("banana")
print(fruits)       # Output: ['apple', 'cherry', 'banana']

#using pop

fruits = ["apple", "banana", "cherry"]

# Removing and returning the last element
last_fruit = fruits.pop()
print(last_fruit)   # Output: 'cherry'
print(fruits)       # Output: ['apple', 'banana']

# Removing and returning the element at index 0
first_fruit = fruits.pop(0)
print(first_fruit)  # Output: 'apple'
print(fruits)       # Output: ['banana']

#clear statement
fruits = ["apple", "banana", "cherry"]

# Clearing the list
fruits.clear()
print(fruits)       # Output: []
```

```python
#QUESTION 5
'''
Sets in Python are an unordered collection of unique elements. They are a powerful tool for tasks that involve membership testing, eliminating duplicates, and set operations like union, intersection, and difference. Below are the key features of sets along with examples of their use:

Key Features of Sets:
Unordered Collection:

Sets are unordered, meaning the elements do not have a specific order, and their positions can change.
Unique Elements:

Sets automatically eliminate duplicate elements. If you try to add a duplicate, it will be ignored.
Mutable:

You can add or remove elements from a set after it is created. However, the elements themselves must be immutable (e.g., numbers, strings, tuples).
Set Operations:

Sets support mathematical set operations like union, intersection, difference, and symmetric difference.
No Indexing or Slicing:

Unlike lists or tuples, you cannot access set elements by index or perform slicing because sets are unordered.
Efficient Membership Testing:

Sets are optimized for membership testing, which means checking whether an element is in the set is very fast.
Examples of Set Operations and Use Cases:
1. Creating a Set:
You can create a set using curly braces {} or the set() function.'''

fruits = {"apple", "banana", "cherry"}
print(fruits)  # Output: {'apple', 'banana', 'cherry'}

# Creating an empty set
empty_set = set()
'''
2. Eliminating Duplicates:
Sets automatically remove duplicate values, which is useful for deduplication tasks.'''

numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = set(numbers)
print(unique_numbers)  # Output: {1, 2, 3, 4, 5}
'''
3. Adding and Removing Elements:
You can add elements using add() and remove them using remove() or discard().'''

fruits = {"apple", "banana", "cherry"}
fruits.add("date")
print(fruits)  # Output: {'apple', 'banana', 'cherry', 'date'}

fruits.remove("banana")
print(fruits)  # Output: {'apple', 'cherry', 'date'}
'''
4. Set Operations:
a) Union (| or union()):

Combines elements from two sets, eliminating duplicates.'''

set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1 | set2
print(union_set)  # Output: {1, 2, 3, 4, 5}
'''
b) Intersection (& or intersection()):

Returns only the elements present in both sets.'''


intersection_set = set1 & set2
print(intersection_set)  # Output: {3}
'''
c) Difference (- or difference()):

Returns the elements present in the first set but not in the second.'''

difference_set = set1 - set2
print(difference_set)  # Output: {1, 2}
'''
d) Symmetric Difference (^ or symmetric_difference()):

Returns elements present in either set, but not in both.'''

symmetric_difference_set = set1 ^ set2
print(symmetric_difference_set)  # Output: {1, 2, 4, 5}
'''
5. Membership Testing:
Sets provide an efficient way to test for membership.'''

fruits = {"apple", "banana", "cherry"}
print("apple" in fruits)  # Output: True
print("date" in fruits)   # Output: False
'''
6. Iterating Through a Set:
You can loop through a set just like other collections.'''

fruits = {"apple", "banana", "cherry"}
for fruit in fruits:
    print(fruit)
# Output could be in any order: 'apple', 'banana', 'cherry'
'''
7. Use Cases of Sets:
a) Removing Duplicates from a List:'''

numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = list(set(numbers))
print(unique_numbers)  # Output: [1, 2, 3, 4, 5]
'''
b) Finding Common Elements Between Two Lists:'''

list1 = [1, 2, 3, 4]
list2 = [3, 4, 5, 6]
common_elements = set(list1) & set(list2)
print(common_elements)  # Output: {3, 4}
```

```python
# QUESTION 7
'''
How to Add, Modify, and Delete Items in a Dictionary in Python
A dictionary in Python is a mutable, unordered collection of key-value pairs. Here's how you can add, modify, and delete items in a dictionary.

1. Adding Items to a Dictionary
To add an item to a dictionary, you assign a value to a new key. If the key already exists, the value associated with it will be updated.
'''

# Creating an empty dictionary
student_grades = {}

# Adding items to the dictionary
student_grades['John'] = 85
student_grades['Alice'] = 92

print(student_grades)
# Output: {'John': 85, 'Alice': 92}
'''
2. Modifying Items in a Dictionary
To modify an item in a dictionary, you simply assign a new value to an existing key.
'''
# Modify an existing value
student_grades['John'] = 88

print(student_grades)
# Output: {'John': 88, 'Alice': 92}
'''
3. Deleting Items from a Dictionary
There are several ways to delete items from a dictionary:
'''
'''
Using del: Deletes a key-value pair from the dictionary.

Using pop(): Removes the item with the specified key and returns its value.

Using popitem(): Removes and returns the last key-value pair added to the dictionary.

Using clear(): Removes all items from the dictionary.
'''

#Using del:

# Delete a specific item
del student_grades['Alice']

print(student_grades)
# Output: {'John': 88}

#Using pop():


# Remove and return the value associated with 'John'
john_grade = student_grades.pop('John')

print(john_grade)
# Output: 88

print(student_grades)
# Output: {}

#Using pop item():

# Add some items back
student_grades['John'] = 88
student_grades['Alice'] = 92

# Remove the last item added (in this case, 'Alice')
last_item = student_grades.popitem()

print(last_item)
# Output: ('Alice', 92)

print(student_grades)
# Output: {'John': 88}

#Using clear():

# Add some items back
student_grades['John'] = 88
student_grades['Alice'] = 92
# Remove all items from the dictionary
student_grades.clear()

print(student_grades)
# Output: {}
```
